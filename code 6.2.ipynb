```python
#1. Create a pie chart presenting the male/female proportion
#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def pie_chart(titanic) :
    pie1 = plt.figure("Pie Chart")
    plt.title("Pie chart for male/female proportion")
    male_count = titanic['sex'].value_counts()['male']
    female_count = titanic['sex'].value_counts()['female']

    labels = 'male', 'female'
    fracs = [male_count, female_count]
    plt.pie(fracs, labels=labels, autopct='%1.1f%%', shadow=True)
    pie1.show()

def scat_plt(titanic) :
    scatter_p = plt.figure("Scatter Plot")
    plt.title("Plot between 'FARE' and 'AGE'")
    plt.xlabel("FARE")
    plt.ylabel("AGE")

    x1 = titanic[titanic['sex'] == 'male']['fare']
    y1 = titanic[titanic['sex'] == 'male']['age']
    x2 = titanic[titanic['sex'] == 'female']['fare']
    y2 = titanic[titanic['sex'] == 'female']['age']
    plt.scatter(x1,y1, c='b', marker='x', label='M', alpha=1)
    plt.scatter(x2,y2, c='r', marker='x', label='F', alpha=1)
    plt.legend()
    scatter_p.show()

    
if __name__ == '__main__':

    url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
    titanic = pd.read_csv(url)
    pie_chart(titanic)
    scat_plt(titanic)
```
